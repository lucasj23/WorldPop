```python
import csv
import matplotlib.pyplot as plt

def read_csv(path):
    data = []
    with open(path, 'r') as csvfile:
        reader = csv.DictReader(csvfile) # No es necesario especificar la coma porque es el delimitador predeterminado en CSV.
        for row in reader:
            data.append(row)
    return data

def population_by_country(data, country):
    result = list(filter(lambda item: item['Country/Territory'] == country, data))  #La función lambda en este caso es simplemente lambda item: item['Country/Territory'] == country, el 'data' es de la funcion filter()
    return result 

#  que hace esta lambda? agarra y le dice que para cada item, analice el valor de la clave [Country/Territory] de ese item (que esperamos que sea un diccionario, para poder tomar su clave) y lo compare con el country del input. 
# Si es coincidente, va a devolver True, y luego el filter lo filtra. 
#item lo creamos internamente para lambda como un parametro y no lo usamos en otro lado. Cuando filter() aplica la funcion lambda a cada elemento de data, internamente se asigna cada elemento a 'item'. 

def get_population(country_data):
    population_dict = {
        '2022': int(country_data['2022 Population']), #si no lo pasamos a entero, matplot lib lo va a graficar mal o no lo va a graficar
        '2020': int(country_data['2020 Population']),
        '2015': int(country_data['2015 Population']),
        '2010': int(country_data['2010 Population']),
        '2000': int(country_data['2000 Population']),
        '1990': int(country_data['1990 Population']),
        '1980': int(country_data['1980 Population']),
        '1970': int(country_data['1970 Population'])
    }
    labels = population_dict.keys()
    values = population_dict.values()
    return labels, values

def generate_bar_chart(labels, values):
    fig, ax = plt.subplots()
    ax.bar(labels, values)
    plt.xticks(rotation=45)
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.title('Population over the years')
    plt.show()

def run():
    data = read_csv('./app/data.csv')
    country = input('Type Country > ')
    result = population_by_country(data, country)

    if len(result) > 0:
        country_data = result[0]
        labels, values = get_population(country_data)
        generate_bar_chart(labels, values)
    else:
        print('Country not found in the data.')

if __name__ == '__main__':
    run()
```

#### 1.  
Python lee y compila todas las definiciones de funciones y clases: Cuando ejecutas un archivo Python, el intérprete de Python primero lee todo el archivo y compila las definiciones de funciones y clases que encuentra. Durante este proceso, Python registra estas definiciones para que puedan ser utilizadas más tarde en el programa.

#### 2. 
Python no ejecuta el código dentro de if __name__ == '__main__': a menos que el archivo se esté ejecutando directamente: La expresión if __name__ == '__main__': se utiliza para determinar si el archivo se está ejecutando como un programa principal o si está siendo importado como un módulo en otro programa. Si el archivo se ejecuta como un programa principal (es decir, no se importa desde otro archivo), Python ejecutará el código dentro de if __name__ == '__main__':.
#### 3. 
Python puede encontrar y ejecutar run() correctamente: En tu ejemplo, aunque run() se define antes de las otras funciones, Python no ejecuta directamente run() cuando lee el archivo. En su lugar, espera a que se cumpla la condición if __name__ == '__main__':, que ocurre después de la definición de las otras funciones. Cuando se cumple esta condición (porque el archivo se está ejecutando directamente), Python ejecuta run(), y en este punto, todas las funciones que run() utiliza ya han sido definidas y pueden ser utilizadas correctamente.

#### En resumen: 
 Python primero procesa todas las definiciones de funciones y clases, luego ejecuta el código dentro de if __name__ == '__main__': solo si el archivo se está ejecutando directamente. Esto significa que Python puede encontrar y ejecutar run() correctamente incluso si se define antes de otras funciones en el archivo.

```python
import csv
import matplotlib.pyplot as plt

def read_csv(path):
    data = []
    with open(path, 'r') as csvfile:
        reader = csv.DictReader(csvfile) # No es necesario especificar la coma porque es el delimitador predeterminado en CSV.
        for row in reader:
            data.append(row)
    return data

def generate_pie_chart(labels, sizes):
    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    plt.title('World Population Percentage in South America')
    plt.show()

def run():
    data = read_csv('./app/data.csv')
    data = list(filter(lambda item: item['Continent'] == 'South America', data))
    countries = list(map(lambda x: x['Country/Territory'], data))
    percentages = list(map(lambda x: x['World Population Percentage'], data))
    generate_pie_chart(countries, percentages)
    
if __name__ == '__main__':
    run()
    
```
