```python
import matplotlib.pyplot as plt
```

```python
def generate_bar_chart(labels, values): 
    fig, ax = plt.subplots()
    ax.bar(labels, values)
    plt.show()

if __name__ == '__main__':  #con esto reconoce que es un script, recordar
    labels = ['a', 'b', 'c']
    values = [100 , 40, 75]
    generate_bar_chart(labels, values)
```

```python
def generate_pie_chart(labels, values): 
    fig, ax = plt.subplots()
    ax.pie(values, labels=labels) # ojo, asi es la sintaxis para hacer pie charts, es una cuestion de la libreria. No se puede cmabiar, solo hay que respetarla. 
    ax.axis('equal') # esto  es para que est√© en el centro
    plt.show()

if __name__ == '__main__':  # con esto reconoce que es un script, recordar
       labels = ['a', 'b', 'c']
       values = [100 , 40, 75]
       generate_pie_chart(labels, values)
```
